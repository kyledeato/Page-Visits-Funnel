```python
import pandas as pd
```

```python
checkout = pd.read_csv('checkout.csv',
                       parse_dates=[1])
checkout.drop_duplicates(['user_id'],inplace=True)
checkout.to_csv('checkout-dedupe.csv',index=False)
```

Import Files

```python
visits = pd.read_csv('visits.csv',
                     parse_dates=[1])
cart = pd.read_csv('cart.csv',
                   parse_dates=[1])
                   
checkout = pd.read_csv('checkout.csv',
                       parse_dates=[1])
purchase = pd.read_csv('purchase.csv',
                       parse_dates=[1])
```

1. Inspect the DataFrames using print and head:

```python
print(visits.head(5))
print(cart.head(5))
print(checkout.head(5))
print(purchase.head(5))
```

2. Combine visits and cart using a left merge.

```python
visits_and_cart = pd.merge(visits, cart, how="left")
print(visits_and_cart)
```

3. How long is your merged DataFrame?

```python
total_visits = len(visits_and_cart)

print(total_visits)
```

4. How many of the timestamps are null for the column cart_time?

```python
null_cart_times = len(visits_and_cart[visits_and_cart.cart_time.isnull()])

print(null_cart_times)
```

5. What percent of users who visited Cool T-Shirts Inc. ended up not placing a t-shirt in their cart?

```python
visited_not_cart = float(null_cart_times) / float(total_visits)

print(visited_not_cart)
```

6. Repeat the left merge for cart and checkout and count null values. What percentage of users put items in their cart, but did not proceed to checkout?

```python
# get dataframe
cart_to_no_checkout_df =  pd.merge(cart, checkout, how="left")
print(cart_to_no_checkout_df)

# get total of users 
cart_to_no_checkout_total = len(cart_to_no_checkout_df)
print(cart_to_no_checkout_total)

# get number of people who put into cart but not checkout
cart_to_no_checkout_number = len(cart_to_no_checkout_df[cart_to_no_checkout_df.checkout_time.isnull()])
print(cart_to_no_checkout_number, 'users had items in cart but did not proceed to checkout')

#get percentage
cart_to_no_checkout_percent = float(cart_to_no_checkout_number) / float(cart_to_no_checkout_total)

print(cart_to_no_checkout_percent, 'percentage of users put items in their cart, but did not proceed to checkout') 
```

7. Merge all four steps of the funnel, in order, using a series of left merges. Save the results to the variable all_data.

Examine the result using print and head.

```python
all_data = visits_and_cart.merge(cart_to_no_checkout_df, how="left").merge(purchase, how="left")
print(all_data.head())
```

```python
# get data containing only records of users who reached the checkout page
reached_checkout = all_data[~all_data.checkout_time.isnull()]
#get data that reached checkout but never purchased
checkout_not_purchase = all_data[(all_data.purchase_time.isnull()) & (~all_data.checkout_time.isnull())]

checkout_not_purchase_percent = float(len(checkout_not_purchase)) / float(len(reached_checkout))
print("% of users who got to checkout but did not purchase:",checkout_not_purchase_percent)
```

```python
print("{} percent of users who visited the page did not add a t-shirt to their cart".format(round(visited_not_cart*100, 2)))
print("{} percent of users who added a t-shirt to their cart did not checkout".format(round(cart_to_no_checkout_percent*100, 2)))
print("{} percent of users who made it to checkout  did not purchase a shirt".format(round( checkout_not_purchase_percent*100, 2)))
```
